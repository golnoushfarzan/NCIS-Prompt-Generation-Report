```python
# YAKE keyword extraction
language = 'en'
max_ngram = 2
dup_threshold = 0.8
dup_function='seqm'
num_keywords = 100
window_size = 1
extractor = yake.KeywordExtractor(lan=language, n=max_ngram, dedupLim=dup_threshold, dedupFunc=dup_function, top=num_keywords, windowsSize=window_size, features=None)
```

```python
all_keywords = []
keywords = [kw[0] for kw in extractor.extract_keywords(sample_text)]
all_keywords.extend(keywords)
```

```python
# Function to get synonyms of a word using NLTK WordNet
def get_synonyms(word):
    synonyms = []
    for syn in wordnet.synsets(word):
        for lemma in syn.lemmas():
            synonyms.append(lemma.name())
    return synonyms

word = "detains"
synonyms = get_synonyms(word)
```

```python
def extract_crimes_from_file(file_path):
    crimes = []

    with open(file_path, 'r') as file:
        for line in file:
            # Split each line by comma to get individual crime types
            line = line.strip()  # Remove any leading/trailing whitespace
            if line:
                crime_types = line.split(',')
                crimes.extend(crime_types)

    return crimes

# Example usage:
file_path = '/content/Underreporting or missreporting of catch - Crime terminologies (2).txt'
crimes = extract_crimes_from_file(file_path)

print("List of crimes:")
for crime in crimes:
    print(crime.strip())  # Print each crime type (strip any extra whitespace)

# Now 'crimes' list contains all the individual crime types extracted from the file
```

```python
# Generate random prompts based on our format
num_prompts = 100
formatted_prompts = []

for _ in range(num_prompts):
    subject = random.choice(subjects)
    consequence = random.choice(synonyms)
    crime = random.choice(crimes2)
    crimes1 = random.choice(crimes)
    specific = random.choice(specifics)

    # Generate two types of prompts: one with "specific + consequence + crimes1" and one with "specific + consequence + crime"
    if random.choice([True, False]):
        prompt = f"{specific} {consequence} {subject} {crimes1}."
    else:
        prompt = f"{specific} {consequence} {subject} {crime}."

    formatted_prompts.append(prompt)

# Print the generated prompts
for prompt in formatted_prompts:
    print(prompt)
```

```python
# Function to write prompts to a text file
def write_prompts_to_file(formatted_prompts, file_path):
    with open(file_path, 'w', encoding='utf-8') as file:
        for prompt in formatted_prompts:
            file.write(prompt + "\n")
```

```python
import spacy
from sentence_transformers import SentenceTransformer, util
from sentence_transformers.util import cos_sim
from sentence_transformers.quantization import quantize_embeddings

model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
def get_score1(prompt, article):
  embedding_1 = model.encode(prompt, convert_to_tensor=True)
  embedding_2 = model.encode(article, convert_to_tensor=True)

  similarity = util.pytorch_cos_sim(embedding_1, embedding_2).numpy()[0]
  return similarity.max()
```

```python
from bs4 import BeautifulSoup
from selenium import webdriver
# from selenium.webdriver.chrome.options import Options

# headless background execution
options = webdriver.ChromeOptions()
options.add_argument("headless")
```

```python
def get_url_text(url : str, save : str = None, file : str = None) -> str:
    '''
    Uses the Selenium WebDriver to scrape all text from the webpage associated to the provided url.

    url : URL address for webpage to be scraped.
    save : Optional argument for saving scraped text as a user specified file type.
    file : Optional argument for naming file with scraped text.

    Webpage text is returned as a string.
    '''

    driver = webdriver.Chrome(options=options)
    driver.get(url)

    page_soup = BeautifulSoup(driver.page_source, 'html.parser')
    p_list = page_soup.find_all("p")

    text = ''

    for p in p_list:
        text += ' ' + p.get_text()

    if save:
        with open(f"{file}.txt", "w") as text_file:
            text_file.write(text)

    return text
```

```python
from numpy import inf       # This will allow us to retrieve all search results.
import pickle
```

```python
def prompt_to_reports(prompt: str,
                      api_key: str,
                      num_results: int = inf,
                      engine: str = 'google_news',
                      hl: str = 'en',
                      gl: str = 'us',
                      save: bool = False,
                      file: str = 'scraped-results.pkl') -> list:
    '''
    Takes in a prompt and returns a list of strings containing the text of the first n search results.

    prompt: The prompt to be searched.
    api_key: The SerpApi API key.
    num_results: Maximum number of search results to retrieve.
    engine: The search engine to use. Defaults to Google News.
    hl: Language to use for search. Defaults to English. For supported languages, see https://serpapi.com/google-languages
    gl: Country to use for search. Defaults to US. For countries supported, see https://serpapi.com/google-countries
    save: Boolean determining whether scraped text should be pickled for later use.
    file: Optional argument for naming pickle file with scraped text.

    Returns scraped text from search results as a list of strings.
    '''

    result_type = {'google': 'organic_results',
                   'google_news': 'news_results'}

    # Initialize SerpApi client
    client = google_news({
        "q": prompt,
        "engine": engine,
        "hl": hl,
        "gl": gl,
        "api_key": api_key
    })

    results_json = client.get_dict()

    results = results_json.get(result_type[engine], [])
    to_scrape = [result['link'] for result in results[:min(num_results, len(results))]]

    texts = []

    for url in to_scrape:
        texts.append(get_url_text(url))

    if save:
        with open(file, "wb") as pickle_file:
            pickle.dump(texts, pickle_file)

    return texts
```

```python
df = pd.read_csv('cleaned_clustered_keywords.csv')
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Create a document-term matrix for the clusters
X_clusters = vectorizer.fit_transform(df['Keyword'])

# Create a document-term matrix for each top keyword
X_top_keywords = vectorizer.transform(top_keywords)

# Calculate cosine similarity between each cluster and each top keyword
similarity_matrix = cosine_similarity(X_clusters, X_top_keywords)

# Print similarity matrix
print("Similarity matrix:")
print(similarity_matrix)
```

```python
# Assign the top keyword to each cluster based on the highest similarity score
assigned_keywords = [top_keywords[np.argmax(row)] for row in similarity_matrix]

# Add the assigned keywords as a new column in the DataFrame
df['assigned_keyword'] = assigned_keywords

# Save the labeled DataFrame to a new CSV file
df.to_csv('labeled_clustered_keywords.csv', index=False)

print("Clusters labeled and saved to 'labeled_clustered_keywordsnew.csv'")
```
